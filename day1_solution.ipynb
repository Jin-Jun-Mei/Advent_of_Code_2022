```python
from pathlib import Path
```

```python
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day1_input_data.txt"
data = open(Path(file_path + file_name), "r").read()
```

```python
elves_lst = data.split("\n\n")
# print(len(elves_lst))
inventory = {}
elve_nr = 1
for elve in elves_lst :
    calories = elve.split("\n")
    try:
        calories = [int(x) for x in calories]
        inventory[elve_nr] = sum(calories)
        elve_nr += 1
    except:
        continue
inventory = dict(sorted(inventory.items(), key=lambda item: item[1],reverse=True))
first_three = dict(list(inventory.items())[:3])
total = sum(first_three.values())
print(inventory)
print(total)
```
