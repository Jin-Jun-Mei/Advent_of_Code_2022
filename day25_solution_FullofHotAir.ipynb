```python
from pathlib import Path
from itertools import zip_longest
import math
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "example.txt"
file_name = "day25_input_data.txt"
with open(Path(file_path + file_name)) as f:
    data = f.read().splitlines()#("\n")
#print(data)
```

```python
brochure = {"0":0, "1":1, "2":2, "-":-1, "=":-2}
brochure_inverse = {v: k for k, v in brochure.items()}
#print(brochure_inverse[0])
```

```python
fuel_lst  = [list(line) for line in data]
nr_lst = list(zip_longest(*map(reversed, fuel_lst), fillvalue=0))##zip the number in the same position, and combine "vertically".
#nr_lst
```

```python
decimal_total = 0
n = len(nr_lst)

for n in range(len(nr_lst)-1,-1,-1):
    #print("===\n",nr_lst[n])
    rst = sum([x if x not in brochure else brochure[x] for x in nr_lst[n]])
    decimal_total += rst * 5**n
    #print(rst)
print(decimal_total)
#print(SNAFU)
```
