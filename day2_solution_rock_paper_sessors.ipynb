```python
rule_1 = {
    "A X": 3 + 1
    ,"A Y": 6 + 2
    ,"A Z": 0 + 3
    ,"B X": 0 + 1
    ,"B Y": 3 + 2
    ,"B Z": 6 + 3
    ,"C X": 6 + 1
    ,"C Y": 0 + 2
    ,"C Z": 3 + 3
}
rule_2 = {
    "A X": 3 + 0
    ,"A Y": 1 + 3
    ,"A Z": 2 + 6
    ,"B X": 1 + 0
    ,"B Y": 2 + 3
    ,"B Z": 3 + 6
    ,"C X": 2 + 0
    ,"C Y": 3 + 3
    ,"C Z": 1 + 6
}
```

```python
from pathlib import Path
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day2_input_data.txt"
```

```python
strategy_guide = open(Path(file_path + file_name), "r").read().split("\n")
total_score1 = 0
total_score2 = 0
for strategy in strategy_guide:
    try: 
        total_score1 += rule_1[strategy]
    except:
        continue
    try:
        total_score2 += rule_2[strategy]
    except:
        continue
print(total_score1)
print(total_score2)
```
