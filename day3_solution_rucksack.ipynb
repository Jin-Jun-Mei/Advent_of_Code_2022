```python
from pathlib import Path
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day3_input_data.txt"
```

```python
rucksack_lst = open(Path(file_path + file_name), "r").read().split("\n")
# print(rucksack_lst)
```

question 1

```python
priorities  = 0
for item in rucksack_lst:
    compartment1 = set(item [:int(len(item)/2)])
    compartment2 = set(item [int(len(item)/2):])
    # print(type(compartment1))
    repeated_items = compartment1.intersection(compartment2)
    for i in repeated_items:
        if i.islower():
            priorities  += ord(i) - 96
        elif i.isupper():
            priorities  += ord(i) -65 + 27
print(priorities)
#print(repeated_lst)
```

question 2

```python
priorities  = 0
for n in range(0,len(rucksack_lst)-3,3):
    repeated_item = set(rucksack_lst[n]).intersection(set(rucksack_lst[n+1]),set(rucksack_lst[n+2])).pop()
    for i in repeated_item:
        if i.islower():
            priorities  += ord(i) - 96
        elif i.isupper():
            priorities  += ord(i) - 65 + 27
print(priorities)
```
