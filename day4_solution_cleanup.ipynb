```python
from pathlib import Path
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day4_input_data.txt"
```

```python
assignment_pairs = open(Path(file_path + file_name))
```

```python
fully_overlap = 0
partially_overlap = 0
for pair in assignment_pairs:
    pair = pair.split(",")
    task1 = pair[0].split("-")
    task1_start = int(task1[0])
    task1_end = int(task1[1])

    task2 = pair[1].split("-")
    task2_start = int(task2[0])
    task2_end = int(task2[1])

    if ((task1_start <= task2_start) and (task1_end >= task2_end) 
    or (task2_start <= task1_start) and (task2_end >= task1_end)):
        fully_overlap += 1

    if ((task1_start <= task2_start <= task1_end ) 
    or (task2_start <= task1_start <= task2_end)):
        partially_overlap += 1
print(fully_overlap)
print(partially_overlap)

```
