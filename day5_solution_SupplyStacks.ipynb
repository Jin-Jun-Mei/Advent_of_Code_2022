```python
import pandas as pd
import re
from pathlib import Path
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day5_input_data.txt"
file = Path(file_path + file_name)
```

```python

df = pd.read_fwf(file, infer_nrows=8,nrows=8,header=None )
stacks = {}
for i in range(9):
    crates = df[i].dropna().tolist()
    crates.reverse()
    stacks[str(i+1)] = crates

#print(stacks)
```

```python
CrateMover = "9000"
CrateMover = "9001"

with open(file, "r") as f:
    procedures = f.readlines()[10:]
    for procedure  in procedures:
        print(procedure)
        quantity, org, dst = re.findall('[0-9]+', procedure)
        print("before moving")
        print(stacks[org])
        print(stacks[dst])
        if CrateMover == "9000":
            for i in range(int(quantity)):
                moving_crate = stacks[org].pop(-1)
                stacks[dst].append(moving_crate)
        if CrateMover == "9001":
            moving_crates = stacks[org][-1*int(quantity):]
            del stacks[org] [-1*int(quantity):]
            stacks[dst] += moving_crates
            print("move ",moving_crates)
        
            # print("move ",crate)
        print("after moving")
        print(stacks[org])
        print(stacks[dst])
        # break
print(stacks)    
for key in stacks.keys():
    print(stacks[key][-1])


#print(intructions)
    
```
