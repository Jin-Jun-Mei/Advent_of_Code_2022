```python
import pandas as pd
import re
from pathlib import Path
file_path = "C:/Users/jinxi/OneDrive - Chalmers/courses/2022_Advent_of_Code/"
file_name = "day6_input_data.txt"
file = Path(file_path + file_name)
```

```python
unique_length = 14 # or 4 for part one 
with open(file) as f:
    datastream  = f.read()
    for i in range(len(datastream) - unique_length):
        characters = datastream [i : i + unique_length]
        # print(len(characters))
        # print(len(set(characters)))
        if (len(set(characters)) == unique_length):
            print(i + unique_length)
            break
        else:
            continue
```
