```python
class Knot:
    instances  = []
    def __init__(self,name, child = None, x = 0, y = 0):
        self.name = name
        self.x, self.y = x, y
        self.child = child
        self.trajectory = [(0,0)]
        Knot.instances .append(self)

    @property
    def x_distance(self):
        return self.x - self.child.x
    
    @property
    def y_distance(self):
        return self.y - self.child.y

    def move_subsequent_knot(self):
        print("\nname",self.name)
        # print("self.trajectory",self.trajectory)
        if self.name != "tail":
            print("self.x,self.y ",self.x,self.y)
            print("child.x, child.y",self.child.x,self.child.y)
            # print("x distance",self.x_distance)
            # print("y distance",(self.x_distance**2 + self.x_distance**2)**(1/2))
            print("distance",(self.x_distance**2 + self.x_distance**2)**(1/2))
            
            if self.x_distance**2 + self.y_distance**2 >  2:
                print("yes move")
                if self.x_distance == 0:##move vertically
                    self.child.y += self.y_distance/ abs(self.y_distance)
                elif self.y_distance == 0:##move horizontally
                    self.child.x += self.x_distance / abs(self.x_distance)
                else: ##move diagonally
                    self.child.x += self.x_distance / abs(self.x_distance)
                    self.child.y += self.y_distance / abs(self.y_distance)
                self.trajectory.append((self.x , self.y))
                self.child.move_subsequent_knot()
            else:
                print("no move")
                self.trajectory.append((self.x , self.y))
                return
        else:
            self.trajectory.append((self.x , self.y))
            return
def generate_knot (first_knot,last_knot):
    if i == last_knot:
        return Knot(name = "tail")
    else:
        return Knot(name=i,child = generate_knot(i+1,last_knot=last_knot))
generate_knot(0,last_knot=1)

H =  [inst for inst in Knot.instances if inst.name == 0][0]
# for knot in Knot.instances :
#     print(knot.name )
```
